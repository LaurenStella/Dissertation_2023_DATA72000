```python
import pandas as pd
import numpy as np
import pickle
from matplotlib import pyplot as plt
```

```python
features_lag_autoreg = ['lat', 'lon', 'Day_of_year', 'Year', 'FLNS', 'FSNS', 'PRECT', 'PRSN',
       'QBOT', 'TREFHT', 'UBOT', 'VBOT', 'Tu_max 1-lag', 'FLNS 1-lag',
       'FSNS 1-lag', 'PRECT 1-lag', 'PRSN 1-lag', 'QBOT 1-lag', 'TREFHT 1-lag',
       'UBOT 1-lag', 'VBOT 1-lag']
```

# CatBoost Feature Importance

## 1-LAGGED-AUTOREG_catboost

```python
# Loading in saved model
with open("1-lagged-autoreg_catboost.pickle", "rb") as f:
    model = pickle.load(f)
    
model
```

```python
# get importance
importance = model.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
    print(features_lag_autoreg[i], ': %.4f' % (v))
# plot feature importance
fig, ax = plt.subplots(figsize=(10,5)) # set the size that you'd like (width, height)
plt.bar(features_lag_autoreg,importance)
plt.xticks(rotation=-70)
plt.xlabel('Feature')
plt.ylabel('Feature importance score')

plt.savefig('1-lagged-autoreg_catboost Feature Importance.png', bbox_inches="tight", pad_inches=0.5)
```
